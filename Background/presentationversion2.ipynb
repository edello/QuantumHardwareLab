## Allgemine Beschreibung

We will be examining a single-qubit system realized through a trapped ion. Further, we want to focus on the differences in performance between a hyperfine and an optical qubit, realized through a [...] and [...] ion, respectively. The goal is then to establish the influence of outer factors, like temperature and magnetic field fluctuations on the performance and coherence times of these systems. While we would generally expect hyperfine qubits to be much more stable in this regard, it is still in our interest to hopefully find a way to qualify and/or quantify these effects. As both systems have generally very different coherence times from the beginning, we want to compare the coherence times using computation times as a point of reference. In other words, how fast and well do certain gates perform with respect to these outer factors on, e.g. the coherence time. Hopefully, this will allow us to gain a more objective view on the comparison results than just comparing certain constants, like T2, directly.

## Kontrolle eines Ionenfallen-Qubits
Ein berblick ber Aufbau, Qubitansteuerung, Kohrenzzeiten und Datenanalyse eines Ca-Ionen-Qubits.


### Aufbau


* **Laserquellen**
  * Khlung + Repumpen: 397 nm, 866 nm, 854 nm
  * Qubit-Laser: **729 nm** (optischer quadrupolarer bergang)
* **Optische Komponenten**
  * Spiegel, Linsen, Polarisatoren, AOMs zur Frequenz-/Intensittssteuerung
* **Strahlteiler / Faserkoppler** fr stabilen Transport des 729-nm-Lichts
* **Detektoren**
  * Photomultiplier (PMT), EMCCD
* **Ultrahochvakuumkammer ~ 10 mbar**
  * Minimiert Kollisionen  lngere Kohrenzzeiten
* **Ionenfalle / Paul Trap**
  * Radiale RF-Konfinierung, axiales DC-Potential
* **Frequenzstabilisierung**
  * Ultra-stabiler Referenzcavity-Laser fr 729 nm


### Khlung


* **Doppler Cooling**  
  * Temperatur ~ 0.11 mK  
  * Limit: $ T_D = \frac{\hbar\Gamma}{2k_B} $
* **Sideband Cooling**
  * Anregung des *red sidebands*: $|n\rangle  |n-1\rangle$
  * Bringt das Ion in den quantisierten Grundzustand  
  * Typisch: $\bar{n} < 0.1$

### Stabilitt des Qubits


* **T1  Energie-Kohrenz**
  * Optische Qubits (729 nm): sehr lange T durch quadrupolaren bergang  
  * Hyperfine Qubits: extrem lange T, spontane Emission ausgeschlossen
* **T2  Phasenkohrenz**
  * meist krzer als T  
  * limitiert durch:
    * Laserfrequenzrauschen
    * Magnetfeldrauschen
    * Temperaturdrift
    * AC-Stark-Effekt
* **Optische Qubits**
  * grere Rabi-Frequenzen  hhere Gate-Fidelity
  * laserrausch-sensitiv
* **Hyperfine Qubits**
  * sehr lange T  
  * T limitiert durch Magnetfelddrift  
  * steuerbar via Mikrowellen oder Raman-Laser
* **Mikrowellenrauschen**
  * meist besser kontrollierbar als optisches Rauschen


### Gates

* **Single-Qubit Operationen**
  * beschreibbar als Rotationen auf der Bloch-Kugel
  * optische Qubits: einzelner Laser (~10 Hz)
  * hyperfeine Qubits: Mikrowellenlaser oder Raman-Laserpaare
* **Rabi-Oszillation**
  $
  P_e(t)=\sin^2\left(\frac{\Omega t}{2}\right)
  $

## Datenanalyse


### Rabi-Frequenz


* Messung der Wahrscheinlichkeit des angeregten Zustands  
* Lsung der Schrdingergleichung zeigt sinusfrmige Oszillation  
* Fitfunktion:
  $
  P_e(t)=\sin^2\left(\frac{\Omega t}{2}\right)e^{-t/T_2^*}
  $
* Parameter:
  * $\Omega$  Rabi-Frequenz  
  * $T_2^*$  Dephasierungszeit  
* Fit via nichtlineare Regression








### Kohrenzzeiten


* Ramsey-Experiment:
  * zwei $\pi/2$-Pulse trennen durch freie Przession
  * Ergebnis:
    $
    P_e(t)=\tfrac12\left(1 + e^{-t/T_2}\cos(\Delta t)\right)
    $
* Spin Echo zur Entfernung langsamer Drifts


### Statistische Unsicherheiten


* **Projection Noise**  
* **Photon Shot Noise**  
* **Fit-Varianz**  

### Systematische Unsicherheiten

* Laserfrequenzdrift  
* Laserintensittsdrift  
* Magnetfelddrift  
* AC-Stark Effekt
